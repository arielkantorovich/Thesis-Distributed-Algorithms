**Some Conclusion:**
There is too much uncertainty between X_feature to Y label We need to try to reduce this uncertainty, or find some relation between feature and labels such that we can learn to predict Y.
1. we can decided that for every Y_label > threshold we map Y_label = ${\inf}$ because lim Pn = 0 in this case.
2. maybe we will try to normalize the input features  using the Z-score and then let's check if there is some correlation between features and labels

First Lets calc Z-score Normalization:
$X_{norm} = ({X - X_{\mu}}) / X_{\sigma}$


```python
X_mean[1], X_mean[4] = 0, 0
X_std[1], X_std[4] = 1, 1
X_norm = (X_train - X_mean) / X_std
```

Plot relation between normalize data and label

```python
plt.figure(1)
plt.figure(figsize=(10, 6))
plt.scatter(X_norm[:, 0], Y_train), plt.xlabel('$g_(n,n)$'), plt.ylabel(r"$\beta_n $")
plt.show()

plt.figure(2)
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_norm[:, 2], Y_train), plt.xlabel('$I_n (for P_n=1)$'), plt.ylabel(r"$\beta_n $")
plt.subplot(1, 2, 2)
plt.scatter(X_norm[:, 3], Y_train), plt.xlabel(r'$\phi_n (for P_n=1)$'), plt.ylabel(r"$\beta_n $"), plt.show()


plt.figure(3)
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_norm[:, 5], Y_train), plt.xlabel('$I_n (for P_n=0.5)$'), plt.ylabel(r"$\beta_n $")
plt.subplot(1, 2, 2)
plt.scatter(X_norm[:, 6], Y_train), plt.xlabel(r'$\phi_n (for P_n=0.5)$'), plt.ylabel(r"$\beta_n $"), plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(X_train[:, 3], X_train[:, 6], Y_train)
ax.set_xlabel(r'$\phi_n1$')
ax.set_ylabel(r'$\phi_n2$')
ax.set_zlabel(r'$\beta_n$')
```
